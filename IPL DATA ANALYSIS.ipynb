```python
#Number of matches played by each team
a['team1'].value_counts().plot(kind='bar')
plt.title('Number of Matches Played by Each Team')
plt.xlabel('Teams')
plt.ylabel('Matches')
plt.xticks(rotation=45)
plt.show()
```

```python
#Match winning percentage by each team
a['match_winner'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Match Wins by Team')
plt.ylabel('')
plt.show()
```

```python
#Matches Played at Each Venue
a['venue'].value_counts().plot(kind='bar', figsize=(10,5))
plt.title('Number of Matches at Each Venue')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=90)
plt.show()
```

```python
#Match Wins by Each Team
a['match_winner'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Matches Won by Each Team')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.xticks(rotation=45)
plt.show()
```

```python
#Toss Decisions Distribution
a['toss_decision'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Toss Decision Distribution')
plt.ylabel('')
plt.show()
```

```python
#Top Players of the Match Awards
a['player_of_the_match'].value_counts().head(10).plot(kind='barh', color='lightgreen')
plt.title('Top 10 Players with Most Player of the Match Awards')
plt.xlabel('Number of Awards')
plt.ylabel('Player')
plt.show()
```

```python
#first_ings_score vs second_ings_score
plt.scatter(a['first_ings_score'], a['second_ings_score'], color='purple')
plt.title('First Innings Score vs Second Innings Score')
plt.xlabel('First Innings Score')
plt.ylabel('Second Innings Score')
plt.grid(True)
plt.show()
```
